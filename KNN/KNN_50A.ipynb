```python
import pandas as pd
import numpy as np
import pickle

file_name = 'dataframe_alpha.xlsx'
file_path = f"{path}{file_name}"

df = pd.read_excel(file_path)
# Filtrar las filas con Age mayor de 40
df_filtrado = df[df['Age'] > 50].reset_index(drop=True)
X = df_filtrado.drop(columns=["Unnamed: 0", "Age"])
y = df_filtrado["Age"]

knn_model = KNNRegressor()
knn_model.set_data(X,y, preprocess=True)

knn_model = KNNRegressor()
knn_model.set_data(X,y)
Plotters = Plotter()
```
